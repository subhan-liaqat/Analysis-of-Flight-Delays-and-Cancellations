```python
import pandas as pd
import psycopg2 as pg
from sqlalchemy import create_engine
```

```python
db_params = {
    'host': 'localhost',
    'database': 'postgres',
    'user': 'postgres',
    'password': 'postgres123'
}
```

```python
conn = pg.connect(
    host = db_params['host'],
    database = db_params['database'],
    user = db_params['user'],
    password = db_params['password']
)

cur = conn.cursor()
conn.set_session(autocommit = True)
```

```python
# Connect to the 'accounts' database

cur.execute("DROP DATABASE IF EXISTS flights")
cur.execute("CREATE DATABASE flights")

db_params['database'] = 'flights'
engine = create_engine(f'postgresql://{db_params['user']}:{db_params['password']}@{db_params['host']}/{db_params['database']}')
```

```python
# Define the file paths for your CSV files

csv_files = {
    'airports': 'dataset/airports.csv',
    'airlines': 'dataset/airlines.csv',
    'flights': 'dataset/flights.csv'
}
```

```python
airports_data = pd.read_csv(csv_files['airports'])
airports_data.head()
```

```python
airlines_data = pd.read_csv(csv_files['airlines'])
airlines_data.head()
```

```python
flights_data = pd.read_csv(csv_files['flights'])
flights_data.head()
```

```python
airports_data.to_sql('airports_data', engine, if_exists='replace', index=False)
airlines_data.to_sql('airlines_data', engine, if_exists='replace', index=False)
flights_data.to_sql('flights_data', engine, if_exists='replace', index=False)
```
